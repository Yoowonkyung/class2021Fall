```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t=np.arange(1,sr*dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
amp = 1; sr = 10000; phase = 0
A  = 440.00
C  = 523.25
F  = 698.46 
D  = 587.33
E  = 659.25	 
F  = 698.46 
G  = 783.99
C2 = 1046.50
```

```python
# Jack Rabbit(산토끼)

notes = [G,E,E,G,E,C,D,E,D,C,E,G,C2,G,C2,G,C2,G,E,G,D,F,E,D,C]

durs = [0.3, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.3, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.3, 0.25, 0.25, 0.25, 0.25, 0.25]
```

```python
s_all = np.empty(0)

for note, dur in zip(notes,durs):
  s, t, th=phasor_sine(amp, sr, dur, note, phase)
  s_all = np. concatenate((s_all,s))
```
